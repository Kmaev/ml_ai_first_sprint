```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tv_series_df = pd.DataFrame(df[df['type'] == 'TV Show'])
```

```python
tv_series_df.drop(['show_id', 'type', 'director', 'duration', 'date_added', 'duration', 'description'], inplace=True, axis=1)
```

```python
tv_series_df.head()
```

```python
tv_series_df.describe(include='all')
```

```python
tv_series_df['cast'] = tv_series_df['cast'].fillna('uninformed cast')
tv_series_df['country'] = tv_series_df['country'].fillna('uninformed_country')
```

```python
tv_series_df['cast'] = tv_series_df['cast'].apply(lambda x:x.split(', '))
tv_series_df['cast'] = tv_series_df['cast'].apply(lambda x:",".join(x[:3]))
```

```python
tv_series_df['listed_in'] =  tv_series_df['listed_in'].apply(lambda x: x.upper())
tv_series_df['cast'] =  tv_series_df['cast'].apply(lambda x: x.upper())
tv_series_df['country'] =  tv_series_df['country'].apply(lambda x: x.upper())
```

```python
tv_series_df.head()
```

```python
tv_series_df['listed_in'] = tv_series_df['listed_in'].str.replace('INTERNATIONAL TV SHOWS,','')
#tv_series_df['listed_in'] = tv_series_df['listed_in'] + ', INTERNATIONAL TV SHOWS'
```

```python
tv_series_df['listed_in'].head()
```

```python
tv_series_df.head()
```

```python
genre_split = tv_series_df['listed_in'].str.split(',', expand=True)
genre_split = genre_split.fillna('-')

cast_split = tv_series_df['cast'].str.split(',', expand=True)
cast_split = cast_split.fillna('-')

genre_split.columns = [f'genre_{i+1}' for i in range(genre_split.shape[1])]
cast_split.columns = [f'cast_{i+1}' for i in range(cast_split.shape[1])]
df_split = pd.concat([tv_series_df['title'], genre_split, cast_split], axis=1)
df_split.fillna('-')
df_split.head()
```

```python
group_dummies = [pd.get_dummies(df_split[y].apply(lambda x: x.strip()), dtype='int') for y in df_split.columns]
group_dummies[0].shape
```

```python
group_dummies[0]
```

```python
group_dummies = pd.concat(group_dummies, axis=1)
group_dummies = group_dummies.fillna(0).astype('uint8')
group_dummies
```

```python
tv_series_df['title'] = tv_series_df['title'].apply(lambda x : x.upper())
```

```python
X_genre_type = np.array(group_dummies)
X_genre_type
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(3,(3*60)), metric='distortion')

visualizer.fit(X_genre_type)        
visualizer.show()
```

```python
X_genre_type.shape
```

```python
kmeans_model = KMeans(n_clusters = 8, random_state=0, algorithm='elkan')

#column_weights = np.ones(X_genre_type.shape[1]) 
#column_weights[53:] = 0.5 

#X_scaled = X_genre_type * column_weights

#y_Kmeans = kmeans_model.fit_predict(X_scaled)
# 
y_Kmeans = kmeans_model.fit_predict(X_genre_type)
print(y_Kmeans)
print(np.unique(y_Kmeans))
print(len(np.unique(y_Kmeans)))
```

```python
df_cluster = tv_series_df.copy()
```

```python
df_cluster['clusters_genre'] = y_Kmeans
```

```python
def check_if_exists(tv_series, _df):
    if tv_series.upper() in _df['title'].values:
        print('The value {} exists in the "type" column.'.format(tv_series))
    else:
        print('The value {} does not exist in the "type" column.'.format(tv_series))
```

```python
from dataclasses import dataclass

@dataclass
class QueryRecommends:
    dataset: pd.DataFrame 
    name: str  
    top_n: int  
    
def recommends(query):
    name_upper = query.name.upper()
    print(name_upper)
    
    title_match = query.dataset[query.dataset['title'] == name_upper]
    cluster_genre = title_match['clusters_genre'].iloc[0]
    recommendations = query.dataset[query.dataset['clusters_genre'] == cluster_genre]
    
    top_recommendations = recommendations[['title', 'listed_in', 'cast',"clusters_genre"]].head(query.top_n)
    print(title_match['listed_in'].iloc[0])
    print(title_match['cast'].iloc[0])
    
    return top_recommendations
```

```python
series_to_check = "squid game"
check_if_exists(tv_series=series_to_check, _df=tv_series_df)
```

```python
query = QueryRecommends(dataset=df_cluster, name=series_to_check, top_n=9)
result = recommends(query)
result
```
