PANDAS PLOTS

```python
ds["Age"].plot(kind="hist", title="Age")
```

```python

ds.Survived.value_counts().plot(kind="pie", labels=["Not Survived", "Survived"], autopct = "%1.1f%%", title="Survival", colors = ["blue", "green"])
```

```python
ds.Pclass.value_counts().plot(kind="barh", title="Passengers Class")
```

```python
ds.Pclass.value_counts().plot(kind="pie",autopct = "%1.1f%%", title="Class Class Distribution", colors = ["red", "blue", "yellow"])
```

```python
ds.Sex.value_counts().plot(kind="pie",autopct = "%1.1f%%", title="Sex", colors = ["blue", "green"])
```

SEABORN PLOTS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds.head()
```

```python
sns.barplot(x=ds.Pclass, y=ds.Survived, data=ds)
plt.title("Survival Based on Class")
```

```python

tc = ds.pivot_table(values="Survived", index = ["Pclass"], columns = ["Sex"])
sns.heatmap(tc, cmap="YlGnBu")
plt.title("Survival Based on Sex and Class")
```

```python
sns.scatterplot(data = ds, x="Fare", y= "Age", hue = "Sex")
plt.title("Age vs Fare, colored by Sex")
```

```python
sns.scatterplot(data = ds, x="PassengerId", y= "Fare", hue = "Embarked")
plt.title("Paid Fare Based on Embarkation Location")
```

```python
sns.lineplot(hue="Sex", x="Pclass", y="Survived", data=ds)
```

```python
sns.set_theme(style="whitegrid")
sns.displot(
    data=ds,
    x="Age", hue="Pclass",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
)
```

```python
sns.set_theme(style="ticks")
sns.histplot(ds, x="Age", hue="Embarked", multiple="stack")
```

MATHPLOTLIB PLOTS

```python
g = sns.FacetGrid(ds, col = "Sex", hue = "Survived", palette = "YlGnBu")
g.map(plt.hist, "Age", bins = 20)
```

```python
g = sns.FacetGrid(ds, col = "Pclass", row = "Sex", hue = "Survived", palette = "YlGnBu")
g.map(plt.hist, "Age", bins = 20)
```
